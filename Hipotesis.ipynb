# COLOMBIA

```python
import pandas as pd
import requests

```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Llamado al archivo excel
data = "/content/Data.xlsx"

#Lectura de la Hoja Colombia
opendata = pd.read_excel(data, sheet_name="Co")

#Selección de Columnas para graficar
colums = opendata[["Fecha", "Consumo GWh"]]

print(opendata)

#Datos de Colombia
```

```python
#Visualzación consumo Colombia

plt.figure(figsize=(10, 6))
plt.plot(colums['Fecha'], colums['Consumo GWh'], marker='o', linestyle='-')

plt.xlabel('Fecha')
plt.ylabel('Consumo GWh')
plt.title('Consumo GWh por años en Colombia')

plt.grid(True)
plt.show()


```

```python
plt.figure(figsize=(10, 6))
plt.bar(colums['Fecha'],colums['Consumo GWh'])

plt.xlabel('Fecha')
plt.ylabel('Consumo GWh')
plt.title('Consumo GWh por años en Colombia')

plt.show()
```

# MEXICO

```python
#Llamado al archivo excel
data = "/content/Data.xlsx"

#Lectura de la Hoja Mexico
data_mx = pd.read_excel(data, sheet_name="Mx")

#Selección de Columnas para graficar
colums_mx = data_mx[["Fecha", "Consumo GWh"]]

print(data_mx)

#Datos de mexico
```

```python
#Visualzación consumo Mexico

plt.figure(figsize=(10, 6))
plt.plot(colums_mx['Fecha'], colums_mx['Consumo GWh'], marker='o', linestyle='-',color='green')

plt.xlabel('Fecha')
plt.ylabel('Consumo GWh')
plt.title('Consumo GWh por años en Mexico')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(colums_mx['Fecha'],colums_mx['Consumo GWh'],color='green')

plt.xlabel('Fecha')
plt.ylabel('Consumo GWh')
plt.title('Consumo GWh por años en Mexico')

plt.show()
```

# Consumo Colombia y Mexico


```python
#Visualizacion entre Colombia y Mexico
plt.figure(figsize=(10, 6))
plt.plot(colums_mx['Fecha'], colums_mx['Consumo GWh'], marker='o', linestyle='-',color='green',label="Mexico")
plt.plot(colums['Fecha'], colums['Consumo GWh'], marker='o', linestyle='-',color='blue',label="Colombia")

plt.xlabel('Fecha')
plt.ylabel('Consumo GWh')
plt.title('Consumo GWh por años en Colombia y Mexico')

plt.legend()

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(colums_mx['Fecha'],colums_mx['Consumo GWh'],color='green',label='Mexico')
plt.bar(colums['Fecha'],colums['Consumo GWh'],color='blue',label='Colombia')

plt.legend()

plt.xlabel('Fecha')
plt.ylabel('Consumo GWh')
plt.title('Consumo GWh por años en Colombia y Mexico')

plt.show()
```
